```python
import numpy as np
import matplotlib.pyplot as plt

# Helper function
def zero_pad(s, length, start):
    assert len(s) <= length, "signal too long!"
    ret_sig = np.zeros(length)
    ret_sig[start:start+len(s)] = s
    return ret_sig
```

```python
s1 = np.array([1, 1, -1, 1, 1])
s2 = np.array([1, 1, 1, -1, -1])

rcv1 = np.array([0.2, 0.2, 1, 1, -1.2, 1, 1, 0.2, -0.2])
rcv2 = np.array([0, 1, 1, -1, 2, 2, 1, -1, -1])
idxs = range(-4, 9)
```

```python
plt.stem(s1)
plt.title('Satellite Signal 1')
plt.xlabel('Time Index [n]')
plt.ylabel('Amplitude of signal s1[n]')
plt.figure()
plt.stem(s2)
plt.title('Satellite Signal 2')
plt.xlabel('Time Index [n]')
plt.ylabel('Amplitude of signal s2[n]')
```

```python
plt.stem(rcv1)
plt.title('Received Signal 1')
plt.xlabel('Time Index [n]')
plt.ylabel('Amplitude of signal r1[n]')
plt.figure()
plt.stem(rcv2)
plt.title('Received Signal 2')
plt.xlabel('Time Index [n]')
plt.ylabel('Amplitude of signal r2[n]')
```

```python
# Calculation of correlations of s1 and s2
corr11 = np.correlate(rcv1,s1, mode='full')

corr12 = np.correlate(rcv1,s2, mode='full')

corr21 = np.correlate(rcv2,s1, mode='full')

corr22 = np.correlate(rcv2,s2, mode='full')
```

```python
plt.stem(idxs, corr11)
plt.title('Correlation of s1 with r1')
plt.xlabel('Time Shift of s1 relative to r1 [k]')
plt.ylabel('Amplitude of corr_r1(s1)[k]')
plt.figure()
plt.stem(idxs, corr12)
plt.title('Correlation of s2 with r1')
plt.xlabel('Time Shift of s2 relative to r1 [k]')
plt.ylabel('Amplitude of corr_r1(s2)[k]')
```

# Notice a peak in the first, but not in the second! So the first signal is present, but the second is not

```python
plt.stem(rcv1)
plt.stem(zero_pad(s1, len(rcv1), 2), linefmt='C1o', markerfmt='C1o')
plt.title('Received Signal 1 plotted with shifted satellite signature')
plt.xlabel('Time Index [n]')
plt.ylabel('Amplitude of signals')
plt.legend(['r1[n]','s1[n-2]'])
```

```python
plt.stem(idxs, corr21)
plt.title('Correlation of s1 with r2')
plt.xlabel('Time Shift of s1 relative to r2 [k]')
plt.ylabel('Amplitude of corr_r2(s1)[k]')
plt.figure()
plt.stem(idxs, corr22)
plt.title('Correlation of s2 with r2')
plt.xlabel('Time Shift of s2 relative to r2 [k]')
plt.ylabel('Amplitude of corr_r2(s2)[k]')
```

# There's a peak in both!

```python
plt.stem(rcv2)
plt.stem(zero_pad(s1, len(rcv2), 1), linefmt='C1o', markerfmt='C1o')
plt.stem(zero_pad(s2, len(rcv2), 4), linefmt='C2o', markerfmt='C2o')
plt.title('Received Signal 2 plotted with shifted satellite signature')
plt.xlabel('Time Index [n]')
plt.ylabel('Amplitude of signals')
plt.legend(['r2[n]','s1[n-1]', 's2[n-4]'])
```
